```python
# Explain Sets with an example of unique trekking participants in Nepal.

trekking_participants = set()

while True:
    name = input("Enter participant name: ")

    if name in trekking_participants:
        print(f"{name} is already registered.")
    else:
        trekking_participants.add(name)
        print(f"{name} has been successfully registered!")

    # asking user to add more participants
    choice = input("Do you want to register another participant? give the answer in yes or no: ")
    if choice.lower() != "yes":
        break

# Display all registered participants
print("\nUnique Trekking Participants in Nepal:")
for participant in trekking_participants:
    print(participant)
```

```python
'''1. (12 Marks) Momo Order Calculator Write a Python program where the user
enters momo types and their prices using a dictionary. Ask how many plates of
each momo they want, then calculate and display the total bill. '''

momo = {} #dictionary to store momo details

momo_types = int(input("How many types of momo do you want to enter? "))


for i in range(momo_types):
    momo_name = input(f"Enter name of momo type {i+1}: ")
    price = float(input(f"Enter price for {momo_name}: "))
    momo[momo_name] = price

# Display menu
print("\nMomo Menu:")
for momos, price in momo.items():
    print(f"{momos} - Rs. {price}")

# ask for momo
total_bill = 0
for momos, price in momo.items():
    quantity = int(input(f"How many plates of {momos} do you want? "))
    total_bill += price * quantity

# Display the total bill
print(f"\nYour total bill is: Rs. {total_bill}")
```

```python
'''3. (12 Marks) Trekking Registration Use a set to store names of registered
trekkers. Keep asking names until the user types ”stop”. At the end, print all
unique participants'''
# creating a set to store trekkers details
trekking_participants = set()

while True:
    name = input("Enter trekker name: ")

    if name.lower() == "stop":
        break


    if name in trekking_participants:
        print(f"{name} is already registered.")
    else:
        trekking_participants.add(name)
        print(f"{name} has been registered successfully.")

# Display
print("\nUnique Trekking Participants in Nepal:")
for participant in trekking_participants:
    print(participant)
```

```python
'''5. (12 Marks) Vegetable Market Pricing Create a dictionary of vegetables with
prices. Keep asking the user to enter vegetable name and quantity until they type
”done”. Calculate and print the total price'''


vegetables = {
    "tomato": 120,
    "potato": 70,
    "onion": 70,
    "cauliflower": 30,
    "mushroom": 3000
}

print("Available Vegetables and Prices (per kg):")
for veg, price in vegetables.items():
    print(f"{veg} - Rs. {price}")

total_price = 0

while True:
    veg_name = input("\nEnter vegetable name or type (done) to finish: ")
    if veg_name.lower() == "done":
        break

    if veg_name in vegetables:
        quantity = float(input(f"Enter quantity of {veg_name} in kg: "))
        total_price += vegetables[veg_name] * quantity
    else:
        print("Vegetable not available.")

print(f"\nYour total bill is: Rs. {total_price}")
```
