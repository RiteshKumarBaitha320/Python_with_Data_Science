```python
'''1. (10 points) Functions– Emergency Blood Bank System (Nepal Red Cross) Write a function match_donor(patient_blood_group, donors_list) that:
• Accepts patient’s blood group and a list of donor records (name, blood_group, phone)
• Returns all compatible donors’ names and phone numbers
• Handles the case where no match is found Hardness: Include compatibility logic (O+, O-, etc.) and efficient searching. Before coding:'''

def match_donor(patient_blood_group, donors_list):
    compatibility = {
        "O-": ["O-"],
        "O+": ["O-", "O+"],
        "A-": ["O-", "A-"],
        "A+": ["O-", "O+", "A-", "A+"],
        "B-": ["O-", "B-"],
        "B+": ["O-", "O+", "B-", "B+"],
        "AB-": ["O-", "A-", "B-", "AB-"],
        "AB+": ["O-", "O+", "A-", "A+", "B-", "B+", "AB-", "AB+"]
    }

    patient_blood_group = patient_blood_group.upper()

    if patient_blood_group not in compatibility:
        return "Invalid patient blood group"

    matched_donors = []

    for donor in donors_list:
        name, blood_group, phone = donor
        if blood_group.upper() in compatibility[patient_blood_group]:
            matched_donors.append((name, phone))

    if not matched_donors:
        return "No compatible donor found"

    return matched_donors

donors = [
    ("Ram", "O+", "981145454"),
    ("Sita", "A-", "9825653245"),
    ("Hari", "B+", "9818802859"),
    ("Gita", "AB+", "986565654")
]

patient_group = "A+"

result = match_donor(patient_group, donors)

print("Compatible Donors:")
print(result)

```
