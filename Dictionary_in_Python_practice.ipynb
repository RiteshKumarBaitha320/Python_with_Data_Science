```python
#Dictionary in Python
student={
    "name":"ritesh",
    "age":22,
    "city":"kathmandu"

}
print(student)
```

```python
print(student["name"])
```

```python
print(student.get("education"))
print(student.get("age"))
```

```python
#update new key
student["education"]="BCA"
print(student)
```

```python
# Udate new key and value
student.update({"contact":981889393})
print(student)
```

```python
student.update({"email":"ritesh123@gmail.com"})
print(student)
```

```python
office={
    # "keys":"values"
    "name":"ritesh",
    "age":22,
    "city":"kathmandu",
    "contact":"985453627"
}

for key , value in office.items():
  print(key,"=",value)
```

```python
# print only keys
for key in office:
  print(f"{key}")
```

```python
#print values olny
for value in office.values():
  print(value)
```

```python

my_details={
    "name":"ritesh",
    "age":22,
    "contact":"9818802859"
}

# print specific item in dictionary
print(my_details.get("name"))
print(my_details["age"])
print("Using for loop print the both keys and values")
for key,value in my_details.items():
  print(key,":",value)
```

```python
# Loop through the dictionary and print each keys and value
my_dict = {
    "name": "ritesh",
    "age": 22,
    "city": "ktm"
}

for key, value in my_dict.items():
    print(f"{key}: {value}")
```

```python
# create a dictionary of 3 friends
friends={
    "friend1":{"name":"ritesh","age":22,"contact":"9864530842"},
    "friends2":{"name":"anish","age":25,"contact":"9818802859"},
    "friends3":{"name":"ramesh","age":27,"contact":"827328434"}
}
for friend, details in friends.items():
  print(friend,":",details)
```

```python
for key , value in friends.items():
  print(key,":",value)
```

```python
for key , value in friends["friend1"].items():
  print(key,":",value)
```

```python
# print only name and age of friends 1
# using for loop
count=0
for key, value in friends["friend1"].items():
  print(key,":",value)

  count+=1
  if count==2:
    break

```

```python
# print details of friend1 and friend2 only
for friend in ["friend1","friends2"]:
  print(f"Details of {friend}:")
  for key, value in friends[friend].items():
    print(key,":",value)
  print()
```
