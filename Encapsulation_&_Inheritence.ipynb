```python
# Encapsulation
# Data Hidding(variables) +data protections(methods-Functions)
# Protects data from misuse
# Accidental changes prevents

# Encapsulation without access control

# ---------------Acess Modifier-----------
# 1.public available------>self.name (everyone can access)
# 2.Protected Members _variable (Intermediate)---->self._balance = balance   # protected(accessible in class and child)
# Private Members __variable (Intermediate)--> self.__balance = balance   # private


# Public acess modifier
class Student():
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def display(self):
    print(f"Name:{self.name} Age:{self.age}")

s1=Student("Ritesh",23)
s1.display()

```

```python
s1.age=24
s1.display()
```

```python
# Protected Modifier
class Student:
  def __init__(self,name,age):
    self._name=name
    self._age=age
  def display(self):
    print(f"Name:{self._name} Age:{self._age}")

  def update(self,name,age):
    self._name=name
    self._age=age

s1=Student("Ritesh",23)
s1.display()
```

```python
# protected modifier
class Account:
    def __init__(self, balance):
        self._balance = balance   # protected

    def show_balance(self):
        print("Balance:", self._balance)

acc = Account(10000)
acc.show_balance()

```

```python
print(acc._balance) #allowed but discouraged
```

```python
# private members (__variable)
class Student():
  def __init__(self,name,age):
    self.__name=name
    self.__age=age
  def display(self):
    print(f"Name:{self.__name} Age:{self.__age}")
s1=Student("Ritesh",23)
s1.display()
```

```python

class BankAccount:
  def __init__(self,name,balance):
    self.name=name
    self.__balance=balance

  def show_balance(self):
    print(f"Hello {self.name} your Balance is:{self.__balance}")

  def deposit(self,amount):
    self.__balance+=amount
    print(f"Your account has been Deposited:{amount} , your New Balance is:{self.__balance}")

  def withdraw(self,amount):
    if amount>self.__balance:
      print("Insufficient Balance")
    else:
      self.__balance-=amount
      print(f"Withdrew={amount} New Balance={self.__balance}")

  # if your data is important , hide it
  # if data is shared ,protected it.

acc=BankAccount("Ritesh",10000)
acc.show_balance()
acc.deposit(5000)
acc.withdraw(20000)

```

```python
# -------------------------------
# STEP 1: Create a class
# -------------------------------
# A class is a blueprint for creating objects.
# Here, BankAccount is a blueprint for bank accounts.

class BankAccount:

    # -------------------------------
    # STEP 2: Constructor method
    # -------------------------------
    # __init__ method runs automatically when an object is created.
    # It is used to initialize (assign) values to variables.

    def __init__(self, name, balance):
        # 'self' refers to the current object

        # Public variable
        # Can be accessed from outside the class
        self.name = name

        # Private variable
        # __balance cannot be accessed directly from outside the class
        # This is encapsulation (data hiding)
        self.__balance = balance


    # -------------------------------
    # STEP 3: Method to show balance
    # -------------------------------
    # This method allows users to see balance safely

    def show_balance(self):
        # Private variable is accessed INSIDE the class
        print("Account Holder Name:", self.name)
        print("Current Balance:", self.__balance)


    # -------------------------------
    # STEP 4: Method to deposit money
    # -------------------------------
    # This method increases account balance

    def deposit(self, amount):

        # Check if the deposit amount is valid
        if amount > 0:
            # Add amount to private balance
            self.__balance = self.__balance + amount

            # Display confirmation message
            print("Deposited Amount:", amount)
            print("Updated Balance:", self.__balance)
        else:
            # If amount is zero or negative
            print("Deposit amount must be positive")


    # -------------------------------
    # STEP 5: Method to withdraw money
    # -------------------------------
    # This method decreases account balance safely

    def withdraw(self, amount):

        # If user tries to withdraw more than available balance
        if amount > self.__balance:
            print("Insufficient Balance")

        # If withdrawal amount is zero or negative
        elif amount <= 0:
            print("Withdrawal amount must be positive")

        else:
            # Deduct amount from balance
            self.__balance = self.__balance - amount

            # Display confirmation
            print("Withdrawn Amount:", amount)
            print("Remaining Balance:", self.__balance)


# ======================================================
# STEP 6: Taking user input to create account
# ======================================================

# Ask user for account holder name
name = input("Enter account holder name: ")

# Ask user for initial balance
# int() is used because input() returns string
balance = int(input("Enter initial balance: "))

# Create object of BankAccount class
# This calls the __init__ method automatically
account = BankAccount(name, balance)


# ======================================================
# STEP 7: Menu-driven program using while loop
# ======================================================

# while True keeps the program running until user chooses Exit
while True:

    print("\n------ BANK MENU ------")
    print("1. Show Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    # Take user's menu choice
    choice = int(input("Enter your choice: "))

    # If user selects option 1
    if choice == 1:
        account.show_balance()

    # If user selects option 2
    elif choice == 2:
        amount = int(input("Enter amount to deposit: "))
        account.deposit(amount)

    # If user selects option 3
    elif choice == 3:
        amount = int(input("Enter amount to withdraw: "))
        account.withdraw(amount)

    # If user selects option 4
    elif choice == 4:
        print("Thank you for using the bank system")
        break   # Exit the loop and stop program

    # If user enters wrong choice
    else:
        print("Invalid choice! Please try again.")
```
