A community health center maintains records of patients visiting daily.
Each patient record contains:
• Patient name
• Age
• Symptoms list
• Consultation fee
The health center wants a modular program using functions to:
1. Add patient records
2. Calculate total daily earnings
3. Identify senior citizens (age ≥ 60)
4. Display patients having a specific symptom
5. Generate a daily summary report

```python
# List to store the records of patients
patient_records = []

# Function to add a patient record
def add_patient_record():
    name = input("Enter the patient's name: ")
    age=int(input("Enter the patient's age: "))

    if age < 0:
        print("Age cannot be negative. Try again.")
        return
    else:
        print("Age is valid")

    symptoms = input("Enter the patient's symptoms:")

    consultation_fee = int(input("Enter the consultation fee: "))
    if consultation_fee < 0:
      print("Consultation fee cannot be negative. Try again.")
      return
    else:
      print("Consultation fee is valid")

    # Add the patient record to the list
    patient = {
        "name": name,
        "age": age,
        "symptoms": symptoms,
        "consultation_fee": consultation_fee
    }
    patient_records.append(patient)
    print(f"{name} is successfully recorded!\n")

# Function to calculate total daily earnings
total_earnings = 0
def total_daily_earnings():
  for patients in patient_records:
     total_earnings += patients['consultant_fee']
  print(f"Total daily earnings: Rs.{total_earnings}\n")

# Function to identify senior citizens
def senior_citizens():
  for patient in patient_records:
    if patient["age"] >= 60:
      print(f"{patient['name']} is a senior citizen.\n")


# Function to display patients having a specific symptom
def patients_with_symptom():
    symptom_search = input("Enter the symptom to search: ").strip().lower()
    matched_patients = [patient["name"] for patient in patient_records
                        if symptom_search in [s.lower() for s in patient["symptoms"]]]

    if matched_patients:
        print(f"Patients with '{symptom_search}':", ", ".join(matched_patients), "\n")
    else:
        print(f"No patients found with symptom '{symptom_search}'.\n")

# Function to generate a daily summary report
def daily_summary():
    print("\n--- Daily Summary Report ---")
    print(f"Total patients today: {len(patient_records)}")
    total_daily_earnings()
    senior_citizens()

    print("Patient Details:")
    for patient in patient_records:
        print(f"Name: {patient['name']}, Age: {patient['age']}, Symptoms: {', '.join(patient['symptoms'])}, Fee: Rs.{patient['consultation_fee']}")
    print("----------------------------\n")

# Main program loop
def main():
    while True:
        print("1. Add patient record")
        print("2. Show total daily earnings")
        print("3. Show senior citizens")
        print("4. Find patients by symptom")
        print("5. Show daily summary report")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ").strip()

        if choice == "1":
            add_patient_record()
        elif choice == "2":
            total_daily_earnings()
        elif choice == "3":
            senior_citizens()
        elif choice == "4":
            patients_with_symptom()
        elif choice == "5":
            daily_summary()
        elif choice == "6":
            print("Exiting program. Have a nice day!")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 6.\n")

# Run the program
if __name__ == "__main__":
    main()
```

```python
'''A training institute conducts multiple courses.
Each student enrolls in a course and has:
• Student ID
• Student Name
• Course Name
• Marks in 3 assessments
Write a Python program using functions only (no global logic except
function calls) to perform the following:
Functional Requirements:
1. Add student records
2. Calculate average marks of each student
3. Assign grade based on average
o ≥ 80 → A
o ≥ 60 → B
o ≥ 40 → C
o < 40 → Fail
4. Display all students who passed
5. Display course-wise topper
6. Generate final report'''


# Function to add student records
def add_student_records():
    students = []
    n = int(input("Enter number of students: "))

    for i in range(n):
        print(f"\nEnter details of student {i+1}")
        student_id = input("Student ID: ")
        name = input("Student Name: ")
        course = input("Course Name: ")

        marks = []
        for j in range(3):
            mark = float(input(f"Enter marks for assessment {j+1}: "))
            marks.append(mark)

        student = {
            "id": student_id,
            "name": name,
            "course": course,
            "marks": marks
        }
        students.append(student)

    return students


# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks)


# Function to assign grade based on average
def assign_grade(avg):
    if avg >= 80:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "Fail"


# Function to display all students who passed
def display_passed_students(students):
    print("\n--- Passed Students ---")
    for student in students:
        avg = calculate_average(student["marks"])
        grade = assign_grade(avg)
        if grade != "Fail":
            print(f"{student['name']} ({student['course']}) - Grade: {grade}")


# Function to display course-wise topper
def course_wise_topper(students):
    toppers = {}

    for student in students:
        avg = calculate_average(student["marks"])
        course = student["course"]

        if course not in toppers or avg > toppers[course]["average"]:
            toppers[course] = {
                "name": student["name"],
                "average": avg
            }

    print("\n--- Course-wise Toppers ---")
    for course, data in toppers.items():
        print(f"Course: {course}, Topper: {data['name']}, Average: {data['average']:.2f}")


# Function to generate final report
def generate_final_report(students):
    print("\n--- Final Report ---")
    for student in students:
        avg = calculate_average(student["marks"])
        grade = assign_grade(avg)

        print(f"""
Student ID   : {student['id']}
Name         : {student['name']}
Course       : {student['course']}
Marks        : {student['marks']}
Average      : {avg:.2f}
Grade        : {grade}
----------------------------
""")


# call main functions
def main():
    students = add_student_records()
    display_passed_students(students)
    course_wise_topper(students)
    generate_final_report(students)


# Program execution starts here
main()

```
