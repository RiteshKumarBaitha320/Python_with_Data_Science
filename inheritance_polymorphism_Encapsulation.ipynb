```python
# Controlled Acess
class BankAccount:
  def __init__(self,balance):
    self.__balance=balance

  def deposit(self,amount):
    if amount>0:
      self.__balance+=amount

  def withdraw(self,amount):
    if amount<=self.__balance:
      self.__balance-=amount
    else:
      print("Insufficient Balance")

  def get_balance(self):
    return self.__balance

acc=BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
print(acc.get_balance())
```

```python
class Account:
  def __init__(self,balance):
    self.__balance=balance

  def withdraw(self,amount):
    raise NotImplementedError("Subclass must be implement abstract method")

class SavingAccount(Account):
  def withdraw(self,amount):
    print("Saving Account withdrawal ")

class CurrentAccount(Account):
  def withdraw(self,amount):
    print("Current Account withdrawal ")

account=[SavingAccount(1000),CurrentAccount(2000)]

for acc in account:
  acc.withdraw(500)
```

```python
class Employee:
    def __init__(self, salary):
        self.__salary = salary   # private variable (encapsulation)

    def get_salary(self):
        return self.__salary

    def calculate_pay(self):
        return self.__salary


class FullTime(Employee):
    def __init__(self, salary, bonus):
        super().__init__(salary)
        self.bonus = bonus

    def calculate_pay(self):
        return self.get_salary() + self.bonus


class PartTime(Employee):
    def __init__(self, hourly_rate, hours):
        super().__init__(hourly_rate)
        self.hours = hours

    def calculate_pay(self):
        return self.get_salary() * self.hours


# Objects
emp1 = FullTime(5000, 2000)
emp2 = PartTime(50, 80)

# Output
print("Full-time pay:", emp1.calculate_pay())
print("Part-time pay:", emp2.calculate_pay())


```

```python
class Employee:
    def __init__(self):
        self.__salary = 0   # private variable

    def set_salary(self, amount):
        self.__salary = amount

    def get_salary(self):
        return self.__salary

    def calculate_pay(self):
        # default pay calculation, can be overridden
        return self.get_salary()


class FullTime(Employee):
    def __init__(self, base_salary, bonus):
        super().__init__()
        self.set_salary(base_salary)  # set encapsulated salary
        self.__bonus = bonus          # private bonus

    def calculate_pay(self):
        # pay = base salary + bonus
        return self.get_salary() + self.__bonus


class PartTime(Employee):
    def __init__(self, hourly_rate, hours):
        super().__init__()
        self.set_salary(hourly_rate)  # store hourly rate in salary
        self.__hours = hours          # private hours

    def calculate_pay(self):
        # pay = hourly_rate * hours
        return self.get_salary() * self.__hours


# Objects
emp1 = FullTime(5000, 1500)
emp2 = PartTime(60, 100)

# Output
print("Full-time pay:", emp1.calculate_pay())
print("Part-time pay:", emp2.calculate_pay())
```
