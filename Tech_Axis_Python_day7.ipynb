```python
# insert(position,element):function that add element in the list at specific position
my_list=[1,2,3,5]
my_list.insert(2,"ram")
print(my_list)
```

```python
my_list=[1,2,3,5]
my_list.insert(3,["Ram",7])
print(my_list)
```

```python
#clear():->it is the function that removes all the elements from the list
my_list=[1,2,3]
my_list.clear()
print(my_list)
```

```python
#copy():-->The copy() creates a new copy of the list
a=[1,2,3,4]
b=a.copy()
print(b)
```

```python
# count():-->it checks number of occuring of a specific element in the list
my_list=[1,3,4,3,4,2,1]
result=my_list.count(3)
print(result)
```

```python
# Count how many 2 are there in my_list using for loop
my_list = [1, 2, 3, 2, 5, 2]

count = 0

for item in my_list:
    if item == 2:
        count += 1

print("Total number of two:", count)
```

```python
# extend()-the extend function appends the elements of the another list
# to the end of the other list
my_list=[1,2,3,4,2]
other_list=[5,6,7]
my_list.extend(other_list)
print(my_list)
```
