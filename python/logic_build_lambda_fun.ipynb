```python
order=lambda amount: "large order" if amount>500 else "small order"
a=int(input("Enter the price of the product:"))
print("The order is ",order(a))
```

```python
check_num=lambda num: "Even num" if num%2==0 else "odd num"
a=int(input("Enter any number:"))
print(check_num(a))
```

```python
salary = [10000, 20000, 30000]
add_bonus = list(map(lambda x: x + x * 0.1, salary))
print("Salary with bonus:", add_bonus)
```

```python
marks=[20,50,43,79,98]
add_marks=list(map(lambda x:( x+5 if x<40 else x),marks))
print(add_marks)

```

```python
gym=[("Ram",17),("Shyam",34),("Raj",23)]
eligible_can=list(filter(lambda x:x[1]>=18,gym))
print(eligible_can)
```

```python
super_market=[("Ghee",1000),("Rice",1500),("Honey",1200)]
expensive=list(filter(lambda x:x[1]>1000,super_market))
print(expensive)
```

```python
data=[("ram",67),("gita",98),("ramesh",45),("hari",32)]
sorted_data=sorted(data,key=lambda x:x[1])
print(sorted_data)
```

```python
# lambda x: x if x>10 else x-2 else x-4
def process(x):
  if x>10:
    return x
  elif x<10:
    return x-2
  else:
    return x-4
print(process(15))
print(process(5))
```

```python
order = [("tuborg", 1000), ("vodka", 200), ("Rum", 3000)]
expensive = list(map(lambda x: (x[0], x[1] * 1.05) if x[1] > 500 else x, order))
expensive.sort(key=lambda x: x[1])

print(expensive)
```
