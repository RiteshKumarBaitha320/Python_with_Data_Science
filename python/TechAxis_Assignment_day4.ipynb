```python
#1.Write a program that takes a user's name and age as input and prints a personalized message using an f-string.
age=int(input("Enter your age:"))
name=input("Enter your name")
print(f"Hello! I am {name} and i am {age} years old")
```

```python
#2.Write a program that takes the radius of a circle as input and calculates its area. Print the result using an f-string. "The area of circle is ..."
radius=float(input("Enter the radius of circle:"))
area=3.14*radius**2
print(f"The area of circle is:{area:.3f}")
```

```python
#3.Create a program that prompts the user for their birth year and calculates their age. Display the result using an f-string.
birth_year=int(input("Enter your birth year:"))
current_year=int(input("Enter the current year:"))
age=current_year-birth_year
print(f"My Birth year is {birth_year} and now i am {age} years old ")
```
