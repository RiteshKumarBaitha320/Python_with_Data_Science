```python
# lambda function:-A lambda function is an anonymous function that can have any number of arguments but only one expression.
def square(x):
  return x*x

square(2)
```

```python
# Lambda functions
square = lambda x: x*x
square(2)
```

```python
# multiply two number and check this number is even or odd using lambda functions
multiply=lambda x,y:x*y
print(multiply(2,3))
if multiply(2,3)%2==0:
  print("even")
else:
  print("odd")
```

```python
# lambda to multiply two numbers
multiply = lambda x, y: x * y

# lambda to check even or odd
check_even_odd = lambda n: "Even" if n % 2 == 0 else "Odd"

# inputs
a = 2
b = 3

# processing
result = multiply(a, b)

# output
print("Multiplication:", result)
print("Number is:", check_even_odd(result))
```

```python
# write a program to add two number and check the given number is even or not usng lambda functions
add=lambda x,y:x+y

check_even_odd=lambda n:"Even Number" if n%2==0 else "Odd Number"
a=int(input("Enter the first number:"))
b=int(input("Enter the second Number:"))

result=add(a,b)

print("The sum of two number is",result)
print("The given number is",check_even_odd(result))
```

```python
# filter:- true value return and remove other value
num=[1,2,3,4,5,6,7]
even=list(filter(lambda x:x%2==0,num))
print(even)
```

```python
# without using lambda fucntions
num = [1, 2, 3, 4, 5, 6, 7]

def is_even(x):
    return x % 2 == 0

even = list(filter(is_even, num))
print(even)
```

```python
# using lambda function to filter out those numbers which are greater than 20 from list
b=[10,15,20,25,9,8,4,1]
gt_num=list(filter(lambda x:x>20,b))
print(gt_num)
```

```python
# sorting numbers using lambda function
s=[2,5,3,1,7,9,0,11]
s.sort(key=lambda x:x)
print(s)
```

```python
student=[("Ram",80),("Hari",54),("Gita",75),("Sita",98)]
sorted_students=sorted(student,key=lambda x:x[1])
print(sorted_students)
```

```python
# write a program using lambda function to print maximum number
max_num=lambda x,y: x if x>y else y

a=int(input("Enter any number:"))
b=int(input("Enter any other number:"))

print("Maximum number is:",max_num(a,b))
```

```python
# write a program to calculate absolute value
abs_num=lambda x:x if x>=0 else -x
a=int(input("Enter any number:"))
print("The absolute number is:",abs_num(a))
```

```python
# Convert Celsius to Fahrenheit
temperature=lambda x:(x*9/5)+32
a=float(input("Enter the temperature in degree celcius:"))
print("The temperature in fahrenheit is:",temperature(a))
```

```python
'''Check Divisible by 3
Lambda function that returns "Yes" if a number is divisible by 3, else "No".'''

number=lambda x: "Yes" if x%3==0 else "No"
a=int(input("Enter the number:"))
print(f"The provided number Complete is divisible by 3 :{number(a)} ")
```
