```python
# 1.Write a program to add two numbers entered by the user.

# 2.Take two numbers and print their subtraction result.

# 3.Ask the user for two numbers and display their multiplication.

# 4.Write a program to divide two numbers and print the quotient.

# 5.Take a number from the user and print its remainder when divided by 5.

# 6.Write a program to calculate exponent of number entered by the user.

a=int(input("Enter first number"))
b=int(input("Enter Second number"))
sum=a+b
sub=a-b
mul=a*b
div=a/b
rem=a%b
exp=a**b
print(f"sum={sum}\nsub={sub}\nmul={mul}\ndiv={div:.3f}\nrem={rem}\nexponent={exp}")
```

```python
# 7.Take the radius of a circle and calculate its area using πr².
radius=int(input("Enter the radius of circle"))
area=3.14*radius**2
print(f"The area of circle is:{area:.3f}")
```

```python
# 8.Ask the user for a number and calculate:
# Square
# Cube


side=float(input("Enter the side "))
area=side**2
area1=6*side**2
volume=side**3
print(f"The area of Square is:{area:.3f}\nThe area of cube is:{area1:.3f}\nThe volume of cube is:{volume:.3f}")
```

```python
# 9.Write a program to calculate the simple interest:
# SI = (P × T × R) / 100
principal=int(input("Enter the principal"))
time=int(input("Enter the time"))
rate=int(input("Enter the rate"))
SI=(principal*time*rate)/100
print(f"The Simple interest is:{SI:.3f}")
```

```python
# 10.Take three numbers and calculate the average using arithmetic operators only.
a=int(input("Enter first number"))
b=int(input("Enter second number"))
c=int(input("Enter third number"))
avg=(a+b+c)/3
print(f"The average number is: {avg:.3f}")
```

```python
# b)String Operations in python
# 1. String Creation
# Take a string input from the user and create a new string by reversing every word individually (e.g., "Hello World" → "olleH dlroW").
text=input("Enter the string")
reversing_string=text[::-1]
print(f"The reversing string is:{reversing_string}")
```

```python
# 2. String Length
# Write a program to count the number of vowels and consonants in a given string.
text=input("Enter the String:")
vowels="aeiouAEIOU"
vowel_count=0
const_count=0
print(f"Total character in text is")

for char in text:
  print(char)
  if char.isalpha():
    if char in vowels:
      vowel_count=vowel_count+1
    else:
      const_count=const_count+1
print(f"Total Vowels are:{vowel_count}")
print(f"Total Consonants are:{const_count}")



```

```python
#string indexing and slicing
#indexing
text=input("Enter any text ")
print(f"The first character of text is:{text[0]}")
print(f"The last character of text is:{text[-1]}")

# Slicing
print(text[0:4])
print(text[0:11:2])  #if a text of character 11 then it prints first char and skip 2 char and then print other character and so on
print(text[4:])
print(text[-10:-1])
print(text[::-1]) # reverse string
print(text[0:11])
print(text[2:10:2])
```

```python
#Function in string
# input()
# upper()
# lower
# capitalized()
# replace()
# strip()
# rstrip()
# lstrip()
# split()
# count()
# escaping character
# join()
# checktype()
#swapcase() (convert lower to upper and upper to lower)

text="python Programming Learner    "
print(text.upper())
print(text.lower())
print(text.capitalize())
print(text.replace("python","java"))
print(text.replace("a","@"))
print(text.strip())
print(text.rstrip("aaaabb"))
print(text.lstrip())
print(text.split())
print(text.count("a"))
print("swap upper to lower and lower to upper:",text.swapcase())
print(",".join(text))
print("Check Types:",type(text))
```
