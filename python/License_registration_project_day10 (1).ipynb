```python
''' The Department of Transport Management (DoTM), Nepal wants a simple system to register applicants for a driving license, check eligibility, and display registered applicants.

RULES (Nepal Context)

Minimum age:

16 → Two-wheeler

18 → Four-wheeler

1. Citizenship number must be unique

Applicant chooses license category:

* A → Two-Wheeler
* B → Four-Wheeler

Program runs until user exits'''

'''
---------menu----------
1.Register
2.Check eligibility criteria
3.display registered applicant
1.Register License
 1.1

'''
```

```python
print("-----------Welcome to Department of Transport Management (DOTM)-----------")

citizenship_list = []   # To store citizenship numbers
two_wheeler_license_holder = {}   # Two wheeler applicants
four_wheeler_license_holder = {}  # Four wheeler applicants

while True:
    print("\n-----------Registration View-----------")
    print("1. Register new licence for Two wheeler")
    print("2. Register new licence for Four wheeler")
    print("3. View all registered Applicants")
    print("4. Exit")
    print("--------------------------------------")

    choice = input("Enter any one option from registration view: ")

    # two wheeler details
    if choice == '1':
        print("\nWelcome to Register new licence for Two wheeler")
        name = input("Enter your full name: ")
        age = int(input("Enter your age: "))
        blood_group = input("Enter your Blood Group: ")
        citizenship_num = input("Enter your citizenship number: ")
        district = input("Enter your district: ")

        if age < 16:
            print(f"Sorry {name}! You are not eligible for Two wheeler licence.")
            continue

        if citizenship_num in citizenship_list:
            print("The Citizenship number is already registered.")
            continue

        citizenship_list.append(citizenship_num)
        two_wheeler_license_holder[citizenship_num] = {
            "name": name,
            "age": age,
            "blood_group": blood_group,
            "citizenship_num": citizenship_num,
            "District": district
        }

        print(f"Hi {name}! You are successfully registered for Two Wheeler Licence.")

    # four wheeler details
    elif choice == '2':
        print("\nWelcome to Register new licence for Four wheeler")
        name = input("Enter your full name: ")
        age = int(input("Enter your age: "))
        blood_group = input("Enter your Blood Group: ")
        citizenship_num = input("Enter your citizenship number: ")
        district = input("Enter your district: ")

        if age < 18:
            print(f"Sorry {name}! You are not eligible for Four wheeler licence.")
            continue

        if citizenship_num in citizenship_list:
            print("The Citizenship number is already registered.")
            continue

        citizenship_list.append(citizenship_num)
        four_wheeler_license_holder[citizenship_num] = {
            "name": name,
            "age": age,
            "blood_group": blood_group,
            "citizenship_num": citizenship_num,
            "District": district
        }

        print(f"Hi {name}! You are successfully registered for Four Wheeler Licence.")

    # Third option:-View applicant details
    elif choice == '3':
        if len(two_wheeler_license_holder) == 0 and len(four_wheeler_license_holder) == 0:

            print("No applicant registered yet.")
        else:
            print("\n-----------Registered Applicant Details-----------")

            if two_wheeler_license_holder:
                print("\nTwo Wheeler Applicants:")
                for ctz in two_wheeler_license_holder:
                    data = two_wheeler_license_holder[ctz]
                    print(f"Name: {data['name']} \t Age: {data['age']} \t"
                          f"Blood Group: {data['blood_group']} \t "
                          f"Citizenship: {data['citizenship_num']} \t "
                          f"District: {data['District']}")

            if four_wheeler_license_holder:
                print("\nFour Wheeler Applicants:")
                for ctz in four_wheeler_license_holder:
                    data = four_wheeler_license_holder[ctz]
                    print(f"Name: {data['name']} \t Age: {data['age']} \t "
                          f"Blood Group: {data['blood_group']} \t "
                          f"Citizenship: {data['citizenship_num']} \t "
                          f"District: {data['District']}")

    # exit
    elif choice == '4':
        print("Thank you for using the License Registration System.")
        break


    else:
        print("Invalid choice. Please enter 1, 2, 3 or 4.")
```
