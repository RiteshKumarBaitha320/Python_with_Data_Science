```python
# 1. A bookstore wants to greet a customer using their name and membership type.
# Write a Python program that takes the customer name and membership (Gold,
# Silver, or Regular) as input and prints the greeting using an f-string while applying
# the appropriate discount

name=input("Enter your name: ")
membership=input("Enter your membership type(Gold,Silver or Regular): ")
if membership.lower()=="gold":
  discount="75% Discount"
  print(f"Congratulation {name}! As a {membership} membership, you get {discount} ")
elif membership.lower()=="Silver":
  discount="50% Discount"
  print(f"Congratulation {name}! As a {membership} membership, you get {discount} ")
else:
  print(f"Congratulation {name}! As a {membership} membership, you get {discount}")

```

```python
total = 0
count = 0

while True:
    marks = int(input("Enter marks (negative number to stop): "))

    # Stop loop when a negative number is entered
    if marks < 0:
        break

    # Skip marks above 90
    if marks > 90:
        continue

    # Add valid marks
    total += marks
    count += 1

# Calculate and print average
if count > 0:
    average = total / count
    print(f"Average of valid marks: {average}")
else:
    print("No valid marks entered.")
```

```python
# 3. Write a program that repeatedly asks the user to enter a password until it matches
# the word “admin123”, and display the number of attempts using f-strings

password=""
attempt=0

while password!="admin123":
  password=input("Enter password")
  attempt+=1
print(f"Password matched! number of attempts is:{attempt}")

```

```python
# 4. A shopping cart contains several item prices. Create a program using a for loop that
# stops reading further prices if any value exceeds 5000 using break, and otherwise
# calculates the total bill
total_bill=0
item=int(input("Enter the number of items: "))
for i in range(item):
  price=float(input(f"Enter the price of {i+1}th items:"))
  #stops of price exceeds 5000
  if price>5000:
    print("price Exceeds 5000.Stop furthur input")
    break
  total_bill+=price
print(f"Total price of the several items={total_bill}")



```

```python
# 5. Write a program that takes the number of hours worked by an employee and displays
# whether the employee worked overtime, standard hours, or under-time based on
# conditional checks.

hour=float(input("Enter the working hours of the employee: "))

if hour>8:
  print("Employee worked overtime")
elif hour==8:
  print("Employee worked Standard hour")
else:
  print("Employee worked under-time")
```

```python
'''6. A system logs ten login attempts. Write a program that uses a for loop to print
each attempt number using an f-string and performs a pass operation during the
fifth attempt'''

for attempt in range(1, 11):
    if attempt == 5:
        pass  # doing nothing for 5th attempt
    print(f"Login attempt number: {attempt}")
```

```python
'''7. A bank wants a simple withdrawal validation system. Write a Python program that
reads the account balance and withdrawal amount, checks if the withdrawal can be
made, and prints the result using an f-string.'''

account_balance=float(input("Enter the account balance: "))
withdrawal_amount=float(input("Enter the withdrawal amount: "))

if withdrawal_amount <= account_balance:
    print(f"Withdrawal of Rs. {withdrawal_amount} is successful. Remaining balance: Rs. {account_balance - withdrawal_amount}")
else:
    print(f"Withdrawal failed! You tried to withdraw Rs. {withdrawal_amount}, but your balance is only Rs. {account_balance}.")
```

```python
'''8. Write a program to display all even numbers from 1 to 50 using a for loop and skip
numbers divisible by 10 using the continue statement.'''
print("The even numbers from 1 to 50 are:")
for i in range(1,51):
  if i%2==0:
    if i%10==0:
      continue
    print(f"{i}")
```

```python
'''9. Create a program that asks a student to enter five subject scores using a for loop.
If any score is negative, terminate the loop immediately using break and display an
“Invalid Score!” message.'''

for i in range(5):
  score=float(input(f"Enter the score of subject {i+1}: "))

  if score<0:
    print("Invalid Score!")
    break
  else:
    print("Valid Score!")
```

```python
'''10. A restaurant wants a billing system that accepts order amounts until the user types
“done”. Write a while loop program that ignores orders below 10 using continue
and prints the final total bill'''
total_bill=0
while True:
  order=input("Enter the order amount:")
  if order.lower()=="done":
    break
  amount=float(order)

  if amount<10:
    continue
  total_bill+=amount

print(f"Total bill amount is:{total_bill}")
```

```python
'''11. Write a program that counts and prints the number of vowels in a user-input string
using a for loop and an f-string for output'''
# Get input from the user
text = input("Enter a string: ")

# Initialize vowel count
vowel_count = 0

# List of vowels
vowels = "aeiouAEIOU"

# Loop through each character
for char in text:
    if char in vowels:
        vowel_count += 1

# Print result using f-string
print(f"The number of vowels in the string is: {vowel_count}")
```
