```python
# TypeError: Student.__init__() takes 2 positional arguments but 3 were given
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Ram", 20)
```

```python
# TypeError: Book.__init__() missing 1 required positional argument: 'price'
class Book:
    def __init__(self, title, price):
        self.title = title
        self.price = price

b = Book("Python")
```

```python
# TypeError: Car.__init__() takes 1 positional argument but 2 were given
class Car:
    def __init__(self):
        print("Car created")

c = Car(2025)
```

```python
# TypeError: Manager.__init__() takes 2 positional arguments but 3 were given
class Employee:
    def __init__(self, salary):
        self.salary = salary

class Manager(Employee):
    def __init__(self, bonus):
        self.bonus = bonus

m = Manager(50000, 10000)
```

```python
# TypeError: Student.__init__() takes 2 positional arguments but 3 were given
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, roll):
        self.roll = roll

s = Student("Hari", 12)
```

```python
# AttributeError: 'Account' object has no attribute '__balance'
class Bank:
    def __init__(self, balance):
        self.__balance = balance

class Account(Bank):
    def __init__(self, balance, interest):
        self.__balance = balance
        self.interest = interest

a = Account(10000, 5)
print(a.__balance)
```
