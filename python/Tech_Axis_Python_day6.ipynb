```python
#Control flow statement
# used in for loop and while loop
# maintain the flow of order or instruction
# 1.break statement
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
i = 1
while i <= 10:
    if i == 7:
        break
    print(i)
    i += 1
```

```python
#2.Continue statement:skip the rest of current iteration and move to next
for num in range(1, 11):
    if num == 7:
        break          # stops whole loop when num == 7
    if num % 2 == 0:
        continue       # skip even numbers
    print(num)
```

```python
for i in range(1,6):
  if i==3:
    continue
  print(i)
```

```python
#3.Pass Satement
for i in range(1,6):
  if i==3:
    pass
  print(i)
```

```python
#4.else block with loop
for i in range(5):
  print(i)
else:
  print("Loop completed")
```

```python
for i in range(5):
  if i==2:
    break
  else:
    print("Loop completed")
```
