```python
# conditional logic in function
def check_result(marks):
    if marks >= 40:
        print("Result: pass")
    else:
        print("Result: fail")

marks = int(input("Enter your marks: "))
check_result(marks)
```

```python
def check_marks(marks):
  if marks>=35:
    return "pass"
  else:
    return "fail"
print(check_marks(25))
```

```python
def check_even(num):
  if num%2==0:
    return "even"
  else:
    return "odd"
print(check_even(5))
print(check_even(6))
```

```python
# dealing with multiple parameters
def add(a,b,c):
  return a+b+c
x=int(input("Enter first number:"))
y=int(input("Enter second number:"))
z=int(input("Enter third number:"))
result=add(x,y,z)
print("Sum=",result)
```

```python
def cal_discount(price,discount_percent):
  discount=price*(discount_percent/100)
  final_price=price-discount
  return final_price
print(cal_discount(1000,10))
```

```python
def total_marks(marks):
  total=0
  for m in marks:
    total+=m   #total=total+m
  return total
marks=[60,70,80]
print("Total:",total_marks(marks))
```
