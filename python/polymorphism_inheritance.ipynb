```python
# polymorphism:-The same method name can behave differently for different objects.

class animal:
  def sound(self):
    print("Animal produce sound")

class dog(animal):
  def sound(self):
    print("The dog barks")

a=animal()
b=dog()
a.sound()
b.sound()
```

```python
def make_sound(animal):
  animal.sound()

make_sound(a)
make_sound(b)
```

```python
class Car:
    def move(self):
        print("Car is moving")

class Bike:
    def move(self):
        print("Bike is moving")

def vehicle(v):
    v.move()   # polymorphism

a = Car()
b = Bike()

vehicle(a)
vehicle(b)
```

```python
# polymorphism with inheritance
class Animal:
  def sound(self):
    print("Animal makes sound")
class dog(Animal):
  def sound(self):
    print("Dog barks")

class cat(Animal):
  def sound(self):
    print("Cat meows")

animals=[dog(),cat()]
for animal in animals:
  animal.sound()

```

```python
class Employee:
  def salary(self):
    print("Employee salary")

class Manager(Employee):
  def salary(self):
    print("Higher salary")

class Intern(Employee):
  def salary(self):
    print("Lower salary")

salaries=[Manager(),Intern()]
for salary in salaries:
  salary.salary()
```

```python
# Constructor
class Employee:
  def __init__(self,name):
    self.name=name

  def salary(self):
    print("Salary is not defined")
class Developer(Employee):
  def __init__(self,name,bonus):
    super().__init__(name)
    self.bonus=bonus

  def salary(self):
    print(self.name, "salary is" , 50000+self.bonus)

class Manager(Employee):
  def __init__(self,name,allowance):
    super().__init__(name)
    self.allowance=allowance

  def salary(self):
    print(self.name, "salary is", 80000+self.allowance)

employee=[Developer("Ritesh",20000),
          Manager("Raj",2000)]

for e in employee:
  e.salary()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def price(self):
        print("Price is not defined")

class Car(Vehicle):
    def __init__(self, brand, cost):
        super().__init__(brand)
        self.cost = cost

    def price(self):
        print(self.brand, "costs is ", self.cost)

class Bike(Vehicle):
    def __init__(self, brand, cost):
        super().__init__(brand)
        self.cost = cost

    def price(self):
        print(self.brand, "costs is ", self.cost)

vehicles = [Car("Car", "High Cost"), Bike("Bike", "Low Cost")]

for v in vehicles:
    v.price()
```
