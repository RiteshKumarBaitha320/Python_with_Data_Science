```python
# perform all the directories of the set
# creating two sets a and b
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
print(f"a={a}\nb={b}")
```

```python
# adding one element in set a
a.add(9)
b.add(1)
print(f"a={a}")
print(f"b={b}")

# Adding Multiple elements
a.update(['a','b'])
print(f"After adding multiple element is set a={a}")

```

```python
# Remove one elements from set
a.remove('a')
print(f"after removing the single elements is set a:{a}")
```

```python
# Discard elements  from the set
a.discard('b')
print(f"a={a}")
```

```python
a.discard(11)
print(f"a={a}")
```

```python
# pop element from the set
b={1, 4, 5, 6, 7, 8, 9}
s=b.pop()
print(f"b={b}")
print(f"The poped element is={s}")
print(f"The remaining elements in set b={b}")
```

```python
# superset
a={1,2,3,21,4}
b={3,2,1}
print(f"The superset of a is  b={a.issuperset(b)}")
```

```python
# Disjoint set
a={1,2,3,21,4}
b={3,2,1}
print(f"a={a}\nb={b}")
x=a.isdisjoint(b)
print(f'The disjoint set of a and b is ={x}')
```
