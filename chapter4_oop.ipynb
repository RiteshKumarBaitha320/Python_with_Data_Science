```python
# class , object
# example1
class Student(): # creating a class name student()
   def __init__ (self,name,age): #constructor
    self.name=name #self is used to store data / instance variable
    self.age=age
   def display(self): #self is required to access the variables
     print(f"Name is {self.name} and age is {self.age}")
```

```python
# now lets create an objects
s1=Student("Ritesh",20)
s2=Student("Ali",19)

# memory
# s1-Student object- name =Ritesh, age=20
```

```python
s1.display()
s2.display()
```

```python
class Mobile():
  def __init__(self,brand,price):
    self.brand=brand
    self.price=price
  def show_price(self):
    print(f"Price={self.price}\n brand={self.brand}")

s1=Mobile("Nokia",120000)
s1.show_price()
```
