```python
# Example dictionary
my_dict = {
    "name": "ritesh",
    "age": 22,
    "city": "ktm"
}

for key, value in my_dict.items():
    print(f"{key}: {value}")
```

```python
friends = {
    "friend1": {"name": "ritesh", "age": 20, "city": "ktm"},
    "friend2": {"name": "ram", "age": 22, "city": "pokhara"},
    "friend3": {"name": "shyam", "age": 21, "city": "biratnagar"}
}

# Optional: print all friends
for friend, details in friends.items():
    print(f"{friend}: {details}")
```

```python
product_catalog = {
    "product1": {"name": "Laptop", "price": 80000, "quantity": 5},
    "product2": {"name": "Phone", "price": 30000, "quantity": 10},
    "product3": {"name": "Headphones", "price": 5000, "quantity": 15}
}

# Optional: print the catalog nicely
for product, details in product_catalog.items():
    print(f"{product}: Name={details['name']}, Price={details['price']}, Quantity={details['quantity']}")
```
