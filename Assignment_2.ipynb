```python
# 1. String Operations– Full Program (10 Marks) Write a Python program that: • Takes a string input from the user. • Displays the following:– Length of the string– First and last character– Middle character(s)– First 3 characters and last 3 characters • Checks whether the substring "py" exists in the string. • Reverses the string without using built-in functions.
#take input from the users
text=input("Enter string:")
print(text)

#Length of string
print(f"Length of string= {len(text)}")

#First and last character
print(f"First character:{text[0]}")
print(f"Last character is:{text[-1]}")

#Middle Character
print(f"Middle character is:{text[len(text)//2]}")

#First three characters
print(f"First three characters are:{text[0:3]}")

#Last three characters
print(f"Last three characters are:{text[-3:]}")

# Checks whether the substring "py" exists in the string
if "py" in text:
    print("The substring 'py' exists in the string.")
else:
    print("The substring 'py' does not exist in the string.")

#reverse the string without bulding function
print(f"Reversed string is:{text[::-1]}")

```

```python
# 2.A user enters two numbers using input(). Write a program to show:
# • Addition • Subtraction • Multiplication • Division • Floor division • Modulus • Exponentiation
# Then convert both numbers into strings and perform:
# 2• Concatenation • Repetition • String comparison (> and <)

a=int(input("Enter first number:"))
b=int(input("Enter Second number:2"))
sum=a+b
sub=a-b
mul=a*b
div=a/b
rem=a%b
exp=a**b
fl_div=a//b
print(f"sum={sum}\nsub={sub}\nmul={mul}\ndiv={div:.3f}\nrem={rem}\nexponent={exp}\nfloor div={fl_div}")

print(f"Addition: {sum}")
print(f"Subtraction: {sub}")
print(f"Multiplication: {mul}")
print(f"Division: {div:.3f}")
print(f"Floor Division: {fl_div}")
print(f"Modulus: {rem}")
print(f"Exponentiation: {exp}")

# Convert numbers to strings
a_str = str(a)
b_str = str(b)

# String operations
concate = a_str + b_str
repeat = a_str * b
comparison_gt = a_str > b_str
comparison_lt = a_str < b_str

print(f"\n\nConcatenation of string : {concate}")
print(f"Repetition of string : {repeat}")
print(f"Is first string > second string? {comparison_gt}")
print(f"Is first string < second string? {comparison_lt}")

```
