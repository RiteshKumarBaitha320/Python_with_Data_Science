```python
# Inheritence:
# Why we use Inheritance
# Reuse code (no repetition)
# Easy maintenance
# Supports real-world relationships
# Helps in OOP design

# reusing and rewriting

# class in python
# 1.Parent class/super class
# 2.child class/sub class

class Human: #parent class
  def eat(self):
    print("I can eat")
  def sleep(self):
    print("I can sleep")

```

```python
class Student(Human): #child class
  def study(self):
    print("I can study")
  def play(self):
    print("I can play")
```

```python
s1=Student() #store class student into s1 object
s1.eat()
s1.sleep()
s1.study()
s1.play()
```

```python
class animal:
  def breathe(self):
    print("Animal can breathe")
  def eat(self):
    print("Animal can eat")
  def sleep(self):
    print("Animal can sleep")
class dog(animal):
  def bark(self):
    print("Dog can bark")

s1=dog()
s1.breathe()
s1.eat()
s1.sleep()
s1.bark()
```

```python
# Constructor:-
class Person:
  def __init__(self,name):
    self.name=name
  def introduce(self):
    print(f"My name is {self.name}")

class Student(Person):
  def study(self):
    print("I am Studying")

s1=Student("Ritesh")
s1.introduce()
s1.study()
```

```python
class Student(Person):
  def __init__(self,name,subject):
    super().__init__(name)
    self.subject=subject
  def study(self):
    print(f"I studied {self.subject} myself")

s1=Student("Ritesh","Python")
s1.introduce()
s1.study()
```

```python
s2=Student("Ramesh","java")
s2.introduce()
s2.study()
```

```python
#
class Vehicle:
  def __init__(self,brand,fueltype):
    self.brand=brand
    self.fueltype=fueltype
  def start(self):
    print(f"{self.fueltype} {self.brand} is available in the {self.brand} showroom")

class model(Vehicle):
  def __init__(self,brand,fueltype,modelname):
    super().__init__(brand,fueltype)
    self.modelname=modelname
  def veh_mod(self):
    print(f"I want {self.brand} {self.modelname} model in {self.fueltype} fuel type")

s1=model("BMW","petrol","s1000")
s1.start()
s1.veh_mod()
```

```python
class bird:
  def sound(self):
    print("Bird produce sound")

class sparrow(bird):
  def sparrow(self):
    print("chew chew")

class sd(bird):
  def sd(self):
    print("sjdak")
s1=sparrow()
s2=sd()
s2.sd()
s1.sparrow()
```
