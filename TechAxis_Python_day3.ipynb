```python
# string strip()
#remove unnecessary characters like whiteSpace characters
#1.Tabs(\t,@,%)
#2.newlines(\n,/n#.%,&)

# Removing WhiteSpace
#Example 1:Removing leading and trailing white space
example= "    Hello,world!  "
stripped_example1=example.strip()
print(f"Original:{example}")
print(f"Stripped:{stripped_example1}")
```

```python
#Example 2:Removing leading and trailing newlines and tabs(\n,\t)
example2= "\n\tHello,world!  "
stripped_example2=example2.strip()
print(f"Original:{example2}")
print(f"Stripped:{stripped_example2}")
```

```python
#Removing specific characters from both ends
name="\n--Ritesh , Kumar!--"
stripped_name=name.strip("-")
print(f"Original:{name}")
print(f"Stripped:{stripped_name}")
```

```python
#Removing specific characters from both ends
name="--Ritesh , Kumar--"
stripped_name=name.strip("-")
print(f"Original:{name}")
print(f"Stripped:{stripped_name}")
```

```python
#Removing specific characters from both ends
name="abcdRitesh , Kumar"
stripped_name=name.strip("abcd")
print(f"Original:{name}")
print(f"Stripped:{stripped_name}")
```

```python
#strip(),rstrip()
#Removing specific characters from left and right ends
name="Ritesh  Kumarabcd"
stripped_name=name.rstrip("abcd")
print(f"Original:{name}")
print(f"Stripped:{stripped_name}")
```
