1. Supermarket Billing System (Control Flow + Loop)
A supermarket wants to create a small Python script that:
• Continuously asks the user to enter product prices.
• If the user enters-1, the loop must stop using break.
• Skip prices that are 0 using continue.
• At the end, the system must print:– Total items counted– Total bill amount– Average price

```python
'''1.Supermarket Billing System (Control Flow + Loop) A supermarket wants to create a small Python script that: • Continuously asks the user to enter product prices. • If the user enters-1, the loop must stop using break. • Skip prices that are 0 using continue. • At the end, the system must print:– Total items counted– Total bill amount– Average price'''
total_bill=0
total_items=0
product=int(input("Enter how many product do you want to buy:"))

while True:
  price=int(input(f"Enter price of the product -1:"))

  if price==-1:   # it compare the price with -1 if price is -1 then breaks the statement
    break
  if price==0:  # it compare the price with 0 if price is 0 then skips and  move ahead
    continue

  total_items+=1
  total_bill+=price

if total_items>0:
  avg_price=total_bill/total_items
else:
  avg_price=0

print(f"The total item is: {total_items}")
print(f"Total bill amount:{total_bill}")
print(f"The average price of the product is:{total_bill/total_items}")

```

```python
#2. Student Search System (Loop + else) Given a list of student names:
# students = ["Anish", "Bikash", "Dawa", "Elina", "Rita"]
# Write a program that: • Takes a name from the user. • Searches using a for loop. • Uses break when the name is found. • Prints “Student Found”. • If the loop finishes without a match, use loop-else to print “Student Not Found”.
students=["Anish","Bikash","Dawa","Elina","Rita"]
name=input("Enter the name of the student:")
for i in students:
  if i==name:   # loop compare the name if the same name is matched then print student found message and breaks the statement
    print("Student Found")
    break

else:
    print("Student is not found")
```

```python
'''3. Bank PIN Authentication (while loop + conditions) Write a program that:
• Gives the user 3 attempts to enter the correct PIN (correct PIN = 4321).
• Decreases attempts each time.
• Shows “Access Granted” when correct.
• Shows “Account Locked” after 3 failed attempts.
• Must use while loop and break.'''

i=0
attempts=3
while i<attempts:
  pin=int(input(f"Enter the pin at attempt {i+1}:"))
  if pin==4321:
    print("Access Granted")
    break
  attempts-=1 # after checking the attempts it decrement one by one
  print(f"you tried incorrect pin at attempt {i+1}")

else:
  print("The account is locked")
```

```python
'''4. Digit Processing (Integer → Loop) User enters an integer.
Write a program that:
• Converts the number to string and loops over each digit.
• Counts odd and even digits separately.
• Prints their sum also.
• Must show:– Total odd digits– Total even digits– Sum of all digits'''
num = input("Enter an integer: ")

odd_count = 0
even_count = 0
sum_digits = 0

for digit in num:
    if digit.isdigit():
        digit = int(digit)

        # Count odd and even
        if digit % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

        # Add to sum
        sum_digits += digit


print(f"Total odd Count: {odd_count}")
print(f"Total even Count: {even_count}")
print(f"Sum of all the digits is: {sum_digits}")
```

```python
# 2. Compare break, continue, and pass with examples and explain when each should be used.
# break:it stops the loop completly

for i in range(1, 6):
    if i == 3:
        break
    print(i)
#continue:it skip iteration when the condition meet
print("\nContinue")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
#pass:it does not perform any actions
print("\nPass")
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```
