```python
#To check whether the given number is even or odd
a=int(input("Enter a number:"))
if a%2==0:
  print("The given number is even")
else:
  print("Odd Number")
```

```python
#range defining condition, check if you are teenage or not
age=int(input("Enter your age:"))
if age>=13 and age<=20:
  print("You are a teenage")
else:
  print("You are not teen age")
```

```python
#if, elif and else
#check if a number is positive ,negative or zero
num=int(input("Enter a number:"))

if num>0:
  print("Positive Number")
elif num<0:
  print("Negative Number")
else:
  print("zero")
```

```python
#Implement the grading condition problem.

# A+ => 90 - 100
# A => 80 - 90
# B => 60 - 80
# C => 40 - 60
# D => <40
a=int(input("Enter your marks:"))

if a>=90 and a<=100:
  print("Grade is A+")
elif a>=80 and a<90:
  print("Grade is A")
elif a>=60 and a<80:
  print("Grade is B")
elif a>=40 and a<60:
  print("Grade is C")
else:
  print("Grade D")

```

```python
# Given the BMI of a person, print the following weight status.

# WeightStatusBelow 18.5Underweight18.5 to 24.9Normal weight25 to 29.9Overweight30 to 34.9Obese35 to 39.9Very obeseAbove 40Extremely obese


BMI = float(input("Enter your BMI: "))

if BMI < 18.5:
    print("Underweight")
elif BMI >= 18.5 and BMI <= 24.9:
    print("Normal weight")
elif BMI >= 25 and BMI <= 29.9:
    print("Overweight")
elif BMI >= 30 and BMI <= 34.9:
    print("Obese")
elif BMI >= 35 and BMI <= 39.9:
    print("Very Obese")
elif BMI >= 40:
    print("Extremely obese")
```
