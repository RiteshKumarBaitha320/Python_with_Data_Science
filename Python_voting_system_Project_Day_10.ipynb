Voting regitration system

```python
print("NP welcome to Election Commission of Nepal")
citizenship_list=[] #list to store the citizeship numbers
voters={} #dictionary to store voter details
while True:
    print("\n------------Menu--------------")
    print("1.Register new voter")
    print("2.View all registered voters")
    print("3.Exit")

    user=input("Enter your choice (1/2/3): ")
    if user== "1":
       name=input("Enter your full name: ")
       age=int(input("Enter your age: "))
       citizenship_number=input("Enter your citizenship number: ")
       district=input("Enter your District: ")


       if age<18:
        print(f"Sorry {name}!,You are not eligible to vote (your age must be 18 or above)")
        continue
       if citizenship_number in citizenship_list:
        print("This citizenship number is already register")
        continue
     #  now storing the data
       citizenship_list.append(citizenship_number)
       voters[citizenship_number]={"name":name,
                                "age":age,
                                "district":district
                                }
       print(f"Voter{name} from {district} registered successfully")

    # second option
    # view all the registered voters
    elif user== "2":
      if len(voters)==0:
        print("No voters registered yet")
      else:
        print("\nRegistered voters list:")

        for citizenship_number in voters:
          print(
              f"Name:{voters[citizenship_number]['name']}\t"
              f"Age:{voters[citizenship_number]['age']}\t"
              f"Citizenship_Number:{[citizenship_number]}\t"
              f"District:{voters[citizenship_number]['district']}\t"
              )
      # third option
      # Exit
    elif user=="3":
      print("Thank you for using the voting registration system")
      break

    else:
      print("Invalid choice.Please enter 1,2or 3 ")


```
